####Graphs between alcochol and speeding


```python
#scatterplot
sns.scatterplot(x="alcohol",y="speeding",data=df)

#Inference:From the plot we can say that speeding is increasing when alcohol level is increased
```

```python
#Lineplot
sns.lineplot(x="alcohol",y="speeding",data=df)
#Inference:From the plot we can say that speed is increasing and decreasing,after a point bin alcohol level speeding is gradually increasing...
```

```python
#Distribution plot
sns.distplot(df["alcohol"])
#Inference:From the given plot we can say that alcohol density is increasing upto 4 and high at points between 4 & 6,then cecreasing from 6
```

```python
#Relational plot
sns.relplot(x="alcohol",y="speeding",data=df,hue="abbrev")
#Inference:From the plot we can say that speeding is increasing when alcohol level is increasing for specific abbrev's
```

```python
#Barplot
sns.barplot(x="alcohol",y="speeding",data=df,ci=None)
#Inference:From the plot we can say that Speeding is increasing when alcohol level is increasing
```

```python
#Countplot
sns.countplot(x="alcohol",data=df)
#Inference:From the plot we can say that alcohol level is constant upto a level after that there is a sudden decrease in the count level and then it's normal
```

```python
#Jointplot
sns.jointplot(x="alcohol",y="speeding",data=df)
#Inference:In the plot the graphs above is alcohol &speeding graphs and the dots are bivariate graphs,from the plot we can say that speeding is increasing when alcohol level is increased
```

```python
#Boxplot
sns.boxplot(x="alcohol",y="speeding",data=df)
#Inference:From the plot we can say that both speeding and alcohol are varying and changing each other
```

```python
#Correlation
corr=df.corr()
corr
```

```python
#Heatmap
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```
